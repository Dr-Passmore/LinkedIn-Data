Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Convert CSV to dataframe

```python
data = 'data\Connections.csv'
df = pd.read_csv(data, skiprows = 3)
df = df.dropna(subset=['Company', 'Position'])
df = df.drop(columns=['Email Address', 'First Name', 'Last Name', 'URL'])
df['Connected On'] = pd.to_datetime(df['Connected On'], format='%d %b %Y')
#df['Position'] = df['Position'].replace('Self Employed', 'Self-employed')
print(df)
```

Visualisation 1: Top 10 Positions

```python
top_10_positions = df['Position'].value_counts().nlargest(10)
plt.bar(top_10_positions.index, top_10_positions.values)
plt.xlabel('Position')
plt.ylabel('Number of Connections')
plt.title('Top 10 Positions in LinkedIn Connections')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Visualisation 2: Top 10 Companies

```python
top_10_positions = df['Company'].value_counts().nlargest(10)
plt.bar(top_10_positions.index, top_10_positions.values)
plt.xlabel('Company')
plt.ylabel('Number of Connections')
plt.title('Top 10 Companies in LinkedIn Connections')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
